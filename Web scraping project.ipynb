```python
url = 'https://www.worldometers.info/coronavirus/'
```

```python
import requests
```

```python
content = requests.get(url).text
```

```python
from bs4 import BeautifulSoup
```

```python
Soup = BeautifulSoup(content,'html.parser')
```

```python
Soup 
```

```python
tb = Soup.find('tbody')
```

```python
tb
```

```python
table_row = tb.find_all('tr')
```

```python
table_row
```

```python
table_row[0]
```

```python
table_row[1]
```

```python
table_row[8]
```

```python
all_data = list()
for item in range(8,len(table_row)):
    country_data = list()
    tr = (table_row[item])
    td = tr.find_all('td')
    for dp in td:   # scrapes each country data
        country_data.append(dp.text)
    
    all_data.append(country_data)   # append each countrydata to a whole data
       
```

```python
all_data[0]
```

```python
# 1:10 14
```

```python
map(lambda x:x[1:10]+ [x[14]], all_data)
```

```python
filtered_data = list(map(lambda x : x[1:10]+ [x[14]], all_data))
```

```python
filtered_data[0]
```

```python
column_names = ['country','total cases','new cases','total death','new death','total recovered',
                'new recovered','active cases','serious', 'population']
```
